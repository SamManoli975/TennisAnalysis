```python
import os
os.environ['KMP_DUPLICATE_LIB_OK'] = 'TRUE'
```

```python
from ultralytics import YOLO

# Load a medium-sized pre-trained YOLOv8 model for better accuracy
model = YOLO('yolov8m.pt')

# Start training with optimized parameters
results = model.train(
    data='tennisball.yaml',
    imgsz=416,  # Increased image size for better detection of small objects
    epochs=20,  # More epochs for better learning
    batch=8,  # Increased batch size for better generalization (adjust based on your GPU memory)
    patience=20,  # Early stopping patience to prevent overfitting
    cos_lr=True,  # Use cosine learning rate scheduler
    mixup=0.2,  # Apply mixup augmentation
    copy_paste=0.2,  # Apply copy-paste augmentation
    degrees=10.0,  # Random rotation augmentation
    scale=0.5,  # Random scaling augmentation
    name='yolov8m_custom_improved'
)
```

```python
#from ultralytics import YOLO
import cv2

# Load your trained model
#model = YOLO('path/to/your/best.pt')  # replace with the path to your trained model

# Open the video file
video_path = "TennisMatch.mp4"  # replace with your video path
cap = cv2.VideoCapture(video_path)

# Get the video's width and height
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output.mp4', fourcc, 30, (width, height))

while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()
    
    if success:
        # Run YOLOv8 inference on the frame
        results = model(frame)
        
        # Visualize the results on the frame
        annotated_frame = results[0].plot()
        
        # Write the frame into the output file
        out.write(annotated_frame)
        
        # Display the annotated frame
        cv2.imshow("YOLOv8 Inference", annotated_frame)
        
        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
out.release()
cv2.destroyAllWindows()
```
